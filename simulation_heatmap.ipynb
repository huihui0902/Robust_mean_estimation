```python
import numpy as np
import scipy
import robust_mean_estimate as rl
import importlib as il
from scipy import special
from numpy import linalg as LA
import matplotlib.pyplot as plt
%matplotlib inline
import mpld3
mpld3.enable_notebook()

fsize = 20
fpad = 10
figsize = (8,4)
fontname = 'fantasy'
```

```python
"heatmap -- Levy -- dense biased -- set data -- Top_k"

il.reload(rl)

trials = 50

eps = 0.05
d = 100

xs = list(np.arange(100, 1200, 100))
ys = list(np.arange(4, 12, 1))

keys = rl.Top_K

model = rl.LevyModel()
noise_model = rl.DenseNoise(-2)
model_params = rl.Params(eps = eps, d = d)

inputfilename = 'data/heatmap-Levy-Top_k.pkl'
plot_heatmap = rl.load_data(model = model, noise_model = noise_model, params = model_params, loss = rl.err, keys = keys)
plot_heatmap.setdata_tofile_heatmap(filename = inputfilename, trials = trials, xs = xs, ys = ys)
print(plot_heatmap.Run.runs)
```

```python
"heatmap -- Levy -- dense biased -- plot -- Top_k"

il.reload(rl)

trials = 50

eps = 0.05
d = 100

xs = list(np.arange(100, 900, 100))
ys = list(np.arange(4, 12, 1))

keys = rl.Top_K

model = rl.LevyModel()
noise_model = rl.DenseNoise(-2)
model_params = rl.Params(eps = eps, d = d)

inputfilename = 'data/heatmap-Levy-Top_k.pkl'
outputfilename = 'figs/heatmap-Levy-Top_k.pdf'

title = 'eps = {eps}, d = {d}. Model is Levy. Noise is constant -2 biased.'.format(eps = eps, d = d)
xlabel = 'm'
ylabel = 'k'

heatmap = rl.plot_data(model = model, noise_model = noise_model, params = model_params, loss = rl.err, keys = keys)

heatmap.plotheatmap_fromfile(yscale = 'log', outputfilename = outputfilename, filename = inputfilename, title = title, xlabel = xlabel, ylabel = ylabel, figsize = figsize, fsize = fsize, fpad = fpad, xs = xs, ys = ys, fontname = fontname)
```
