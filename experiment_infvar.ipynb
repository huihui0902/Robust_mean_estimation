```python
import numpy as np
import scipy
import robust_mean_estimate as rl
import importlib as il
from scipy import special
from numpy import linalg as LA
import matplotlib.pyplot as plt
%matplotlib inline
import mpld3
mpld3.enable_notebook()

fsize = 20
fpad = 10
figsize = (8,4)
fontname = 'fantasy'
```

```python
"Top_K -- loss vs param -- T -- dense biased"

il.reload(rl)

trials = 50

m, k, eps, d = 600, 4, 0.1, 100
xs = list(np.arange(1.1, 3.1, 0.01))

keys = [rl.Top_K, rl.Topk_GD, rl.GD_nonsparse, rl.Oracle, rl.RME_sp, rl.RME_sp_npre, rl.Top_K_Filtered]

model = rl.TModel()
noise_model = rl.DenseNoise(-2)
model_params = rl.Params(m = m, k = k, eps = eps, d = d)

inputfilename = 'Data/loss-vs-param-T.pkl'
outputfilename = 'Figs/loss-vs-param-T.pdf'

title = 'Student t'
xlabel = 'param'
ylabel = 'L2 error'

set_data = rl.load_data(model = model, noise_model = noise_model, params = model_params, loss = rl.err, keys = keys)
set_data.setdata_tofile(filename = inputfilename, xvar_name = 'param', trials = trials, xs = xs)

plot = rl.plot_data(model = model, noise_model = noise_model, params = model_params, loss = rl.err, keys = keys)
plot.plotxy_fromfile_time(outputfilename = outputfilename, filename = inputfilename, title = title, xlabel = xlabel, ylabel = ylabel, figsize = figsize, fsize = fsize, fpad = fpad, xs = xs, fontname = fontname)
```
