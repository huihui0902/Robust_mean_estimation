```python
import numpy as np
import scipy
import robust_mean_estimate_acc as rl
import importlib as il
from scipy import special
from numpy import linalg as LA
import matplotlib.pyplot as plt
%matplotlib inline
import mpld3
mpld3.enable_notebook()

fsize = 20
fpad = 10
figsize = (8,4)
fontname = 'Arial'
```

```python
"acc vs eps -- Fisk -- dense biased -- set data"

il.reload(rl)

trials = 50

d, k, m, param= 100, 10, 600, 2
xs = list(np.arange(0.05, 0.5, 0.05))

keys = [rl.Top_K]

model = rl.FiskModel()
noise_model = rl.DenseNoise(-2)
model_params = rl.Params(d = d, k = k, m = m, param = param)

inputfilename = 'data/acc-vs-eps-Fisk.pkl'
plot_m = rl.load_data(model = model, noise_model = noise_model, params = model_params, loss = rl.acc, keys = keys)
plot_m.setdata_tofile(filename = inputfilename, xvar_name = 'acc', trials = trials, xs = xs)
```

```python
"loss vs eps -- Fisk -- dense biased -- plot"

il.reload(rl)

trials = 50

d, k, m, param= 100, 10, 600, 2
xs = list(np.arange(0.05, 0.5, 0.05))

keys = [rl.Top_K]

model = rl.FiskModel()
noise_model = rl.DenseNoise(-2)
model_params = rl.Params(d = d, k = k, m = m, param = param)
inputfilename = 'data/acc-vs-eps-Fisk.pkl'
outputfilename = 'figs/acc-vs-eps-Fisk.pdf'

title = 'Fisk'
xlabel = '$\epsilon$'
ylabel = 'Acc'

plot_eps_loss = rl.plot_data(model = model, noise_model = noise_model, params = model_params, loss = rl.err, keys = keys)

plot_eps_loss.plotxy_fromfile(outputfilename = outputfilename, filename = inputfilename, title = title, xlabel = xlabel, ylabel = ylabel, figsize = figsize, fsize = fsize, fpad = fpad, xs = xs, fontname = fontname)
```
